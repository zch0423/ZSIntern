```python
import sys
sys.path.append("..")
from tools import *
from datetime import datetime
import tushare as ts 
token = "a7c1312fe8187bff67ada6333c4b447caff18f58be66ac8b64d20d5a"
api = ts.pro_api(token)
roePath = "../Mar04/111家roe数据.csv"
markPath = "../Mar04/both.csv"
roe = pd.read_csv(roePath)
mark = pd.read_csv(markPath)
```

```python
def getMoreInfo(code, api):
    '''
    Description:
    返回2017-2020的 roe pe数据
    ---
    Params:
    code, str, like 002010.SZ
    api, tushare api
    ---
    Returns:
    roeInfo, peInfo, DataFrame
    '''
    dates = [f"{year}{month:02d}01" for year in range(2017, 2021) for month in range(1, 13) ]
    results = []
    for date in dates:
        df = api.daily_basic(ts_code=code, trade_date=date, fields='ts_code,trade_date,pe,pe_ttm')
        results.append(df)
    peInfo = pd.concat(results)
    # 业绩快报 yoy同比增长 归属母公司股东的净利润
    roeInfo = api.fina_indicator(ts_code=code, start_date='20170131', end_date='20201231', fields='ts_code,ann_date,end_date,roe,netprofit_yoy')
    return roeInfo, peInfo
```
