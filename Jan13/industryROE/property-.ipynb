```python
def correlation(data):
    values = pd.DataFrame(data.values)
    dataframe = pd.concat([values.shift(5),values.shift(4),values.shift(3),values.shift(2), values.shift(1), values], axis=1)
    dataframe.columns = ["t-5","t-4","t-3","t-2",'t-1', 't']
    result = dataframe.corr()
    return result
```

```python
def drawRolling(d, n=3, wintype=None, save=False, center=False):
    if wintype:
        y1 = d.rolling(n, win_type=wintype, center=center).mean(std=np.std(d.values,ddof=1))
    else:
        y1 = d.rolling(n,center=center).mean()
    fig, axs = plt.subplots(figsize=(14, 6))
    axs.plot(d.index, d.values, label="raw")
    axs.plot(y1.index, y1.values, label="rolling")
    fig.legend()
    if save:
        fig.savefig(save, dpi=400)
```

```python
for company in companies[:5]:
    print(company)
    drawRolling(tt[company], 3, center=False)
```

```python
def getExtreme(data):
    # tt是时间为index的数据
    results = []
    for company in data.columns:
        temp = data[company]
        temp.rolling(3).mean()
        maximum = signal.argrelextrema(temp.values, np.greater)
        minimum = signal.argrelextrema(-temp.values, np.greater)    
        # xx.name为company
        results.append([temp.iloc[maximum], temp.iloc[minimum]])
    return results
```

```python
results = getExtreme(tt)  # 所有的极值点 每个元素表示一个公司，每个元素的第一个表示极大，第二个表示极小
```
