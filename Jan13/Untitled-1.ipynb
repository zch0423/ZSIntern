```python
def drawOne(data1, data3, dataM, dataA, rowIndexes,names=[], drawBool = [1,1,1,1]):
    fig = plt.figure(figsize=(16,8))
    x = [i for i in range(2000, 2021)]
    count = 0
    for rowIndex in rowIndexes:
        if drawBool[0] ==1:
            plt.plot(x, data1.iloc[rowIndex ,2:], label="S1"+names[count])
        if drawBool[1]==1:
            plt.plot(x, data3.iloc[rowIndex, 2:], label="S3"+names[count])
        if drawBool[2]==1:
            plt.plot(x, dataM.iloc[rowIndex, 2:], label="Mid"+names[count])
        if drawBool[3]==1:
            plt.plot(x, dataA.iloc[rowIndex, 2:], label="Annual"+names[count])
        count += 1
    plt.legend()
    plt.show()
```

```python
# 2479 格力地产
def drawOne(data1, data3, dataM, dataA, rowIndexes,names=["", "", "", ""], drawBool = [1,1,1,1],legendBool=False):
    fig = plt.figure(figsize=(16,8))
    x = [i for i in range(2000, 2021)]
    count = 0
    for rowIndex in rowIndexes:
        if drawBool[0] ==1:
            plt.plot(x, data1.iloc[rowIndex ,2:], label="S1 "+names[count])
        if drawBool[1]==1:
            plt.plot(x, data3.iloc[rowIndex, 2:], label="S3 "+names[count])
        if drawBool[2]==1:
            plt.plot(x, dataM.iloc[rowIndex, 2:], label="Mid "+names[count])
        if drawBool[3]==1:
            plt.plot(x, dataA.iloc[rowIndex, 2:], label="Annual "+names[count])
        count += 1
    if legendBool:
        plt.legend()
    plt.show()
```

```python
names = ["wanke", "zhongnan"]  # 万科 中南建设
drawOne(data1, data3, dataM, dataA, rowIndexes = [3080, 3645], names=names, drawBool=[0,0,0,1])
```

```python
properties = "万科A、金地集团、中南建设、金科股份、龙光集团、天健集团、城投控股、阳光城、南山控股".split("、")
properties
```

```python
def selector(x):
    s = ['万科A', '金地集团', '中南建设', '金科股份', '龙光集团', '天健集团', '城投控股', '阳光城', '南山控股']
    for each in s:
        if each in x:
            return True
    return False
```

```python
propertiesIndex = data[data["name"].apply(lambda x:selector(x))].index.values
```
